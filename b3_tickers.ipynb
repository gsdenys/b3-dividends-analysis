```python
from datetime import datetime

def get_historical_data(ticker):
    try:
        stock = yf.Ticker(ticker + '.SA')
        end_date = datetime.now()
        start_date = end_date.replace(year=end_date.year - 3)
        hist = stock.history(start=start_date, end=end_date)
        
        # Calcular a média do preço de fechamento para cada ano
        hist['Year'] = hist.index.year
        avg_close_prices = hist.groupby('Year')['Close'].mean()
        
        # Obter dividendos e calcular DY
        dividends = stock.dividends
        dividends['Year'] = dividends.index.year
        total_dividends = dividends.groupby('Year').sum()
        
        dy = total_dividends / avg_close_prices
        
        # Formatar os resultados em um dict
        result = {}
        for year in range(end_date.year - 3, end_date.year):
            result[f'Avg Close {year}'] = avg_close_prices.get(year, None)
            result[f'DY {year}'] = dy.get(year, None)
            
        return pd.Series(result)
    except Exception as e:
        years = range(datetime.now().year - 3, datetime.now().year)
        avg_close_series = pd.Series({f'Avg Close {year}': None for year in years})
        dy_series = pd.Series({f'DY {year}': None for year in years})
        return pd.concat([avg_close_series, dy_series])

# Aplicar a função para obter dados históricos e adicionar ao DataFrame
historical_data = df['Ticker'].apply(get_historical_data)
df = pd.concat([df, historical_data], axis=1)
```
